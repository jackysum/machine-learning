```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Download and prepare the data
lifesat = pd.read_csv("../Datasets/lifesat.csv")

x = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values
```

```python
# Visualize the data
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()
```

```python
# Select a linear model (model based learning)
model = LinearRegression()

# Train the model
model.fit(x, y)

# Make a prediction for Cyprus
cyprus_gdp = [[37_655.2]]
print(model.predict(cyprus_gdp))

# Visualize the data
gdp_prediction = model.predict(x)

lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.plot(x, gdp_prediction, color="black", label="Linear Regression")
plt.legend()
plt.axis([23_500, 62_500, 4, 9])
plt.show()
```

```python
# Select k-nearest neighbors regression (instance-based learning)
model = KNeighborsRegressor(n_neighbors=3)

# Train the model
model.fit(x, y)

# Make a prediction for Cyprus
cyprus_gdp = [[37_655.2]]
print(model.predict(cyprus_gdp))

# Visualize the data
gdp_prediction = model.predict(x)

lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.plot(x, gdp_prediction, color="black", label="K-Nearest Neighbors")
plt.legend()
plt.axis([23_500, 62_500, 4, 9])
plt.show()
```
